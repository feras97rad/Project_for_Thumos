```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import phik
from phik.report import plot_correlation_matrix
from phik import resources, report
from skimpy import clean_columns
```

```python
# Function for primary analysis of the dataset
def analyze_df(df):
    # Output the shape of the dataframe
    display("Размерность датафрейма:")
    display(df.shape)
    display("")

    # Output the information about the dataframe
    display("Информация о датафрейме:")
    display(df.info())
    display("")

    # Output statistical information about the dataframe
    display("Статистическая информация о датафрейме:")
    display(df.describe())
    display("")

    # output the first 10 lines of the dataframe
    display("Первые 10 строк датафрейма:")
    display(df.head(10))

    # output the number of passes
    display("Кол-во пропусков")
    display(df.isna().sum())
    
    # output the number of duplicates
    display("Кол-во дубликатов")
    display(df.duplicated().sum())
```

# Preparing the dataset

```python
# Cleaning columns names
data = clean_columns(data)

# Checking
data.columns
```

```python
# Replacing str values by numbers in "result" colomn
data['result'] = data['result'].replace({
                                        '5. Purchased Membership': '5',
                                        '4.Purchased Event': '4',
                                        '3. Approved': '3',
                                        'Out of Sequence': '3', 
                                        '2. Sent Personal f/u': '2',
                                        '2021 Member': '2'
                                        })

data['result'] = data['result'].astype('int')

# Checking
data['result'].unique()
```

```python
data.columns
```

```python
# Dropping columns we don't need (like E-mail, full name, etc)
data = data.drop([
    'linkedin_profile', 
    'email', 
    'description',
    'first_name',
    'last_name',
    'full_name', 
    'user_id',
    'job_description', 
    'school_url', 
    'school_url_2', 
    'phone_number_from_drop_contact', 
    'website_from_drop_contact', 
    'twitter', 
    'website', 
    'company_website', 
    'base_url', 
    'profile_id', 
    'timestamp', 
    'mail', 
    'school_description_2', 
    'naf_5_code_from_drop_contact', 
    'naf_5_des_from_drop_contact', 
    'siren_from_drop_contact', 
    'siret_from_drop_contact', 
    'siret_address_from_drop_contact', 
    'siret_zip_from_drop_contact', 
    'vat_from_drop_contact', 
    'school_description', 
    'nb_employees_from_drop_contact', 
    'unnamed_72', 
    'unnamed_73'
], axis=1)

# Checking
data.columns
```

```python
data[['result', 'subscribers', 'connections_count']] = data[['result', 'subscribers', 'connections_count']].astype('Int64')
```

- Based on the correlation matrix, the result depends on headline, mail from drop contact, company, company 2, job description, school date range, birthday and skills.
- Phik correlation was used due to the fact that there are not so many numerical parameters in the dataset 
- Perhaps this is an event for a specific field of activity.

- Most of those who bought tickets and membership have a common contact: Eric Sullivan
- The majority of those who bought tickets and membership have a qualification nominative@pro. Twice as much as catch-all@pro.
- Everyone who bought tickets or membership has a civility "Mr".
- Perhaps this is an event men are more intrested in than women.
- Buying tickets directly depends on the number of contacts in LinkedIN.
- The dataset contains a lot of omissions (NaNs)
- And even more personal data that does not affect the purchase (names, last names,tel. numbers, e-mails, and other contact info)
- Due to the last two factors, it is not possible to apply machine learning without a very long data preprocessing, for which, unfortunately, I did not have enough time. And even at the same time, a satisfying result is not guaranteed.
